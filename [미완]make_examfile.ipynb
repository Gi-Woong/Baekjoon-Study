```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.acmicpc.net/step"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
table = soup.find("table")
```

```python
table_a = table.find_all(lambda x: x.name == "a" and x.attrs.get("href"))
table_a__hrefs = list(map(lambda x: "https://www.acmicpc.net"+x.attrs["href"], table_a))# 링크
table_a__gettext = list(map(lambda x: x.get_text(), table_a))# 폴더명
```

```python
df_table = pd.read_html(str(table))[0].dropna(axis=1)
df_table
```

```python
res = requests.get(list(table_a__hrefs)[0])
soup = BeautifulSoup(res.content, "html.parser")
table = soup.find("table")
df_StepTable = pd.read_html(str(table))[0]
#표가 이상해서 나누고 다시 합치기
df_StepTable = pd.concat([
    df_StepTable[::2].reset_index(drop=True), 
    df_StepTable[1::2]["정보"].reset_index(drop=True)], axis=1)
df_StepTable
```

```python
list_FileName = []
for i, num in enumerate(df_StepTable["문제 번호"]):
    name = f"{str(i+1).zfill(2)}_{num}"
    list_FileName.append(name)
list_FileName
```

```python
df_StepTable["정보"].iloc[:,-1]
```

```python
for i in range(len(df_StepTable)):
    title = df_StepTable["제목"][i]
    info = df_StepTable["정보"].iloc[:,-1][i]
    annot = f"#제목: {title}\n#정보: {info}"
    print(annot)
```

```python
for i, num in enumerate(df_StepTable["문제 번호"]):
    #파일명 지정
    filename = f"{str(i+1).zfill(2)}_{num}"
    #주석 지정
    title = df_StepTable["제목"][i]
    info = df_StepTable["정보"].iloc[:,-1][i]
    annot = f"#제목: {title}\n#정보: {info}"
    #파일 쓰기
    with open(f"./{filename}.py", "w", encoding="utf-8") as w:
        w.write(annot)
        w.close()
```
